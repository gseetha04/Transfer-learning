```python
import numpy as np
import matplotlib.pyplot as plt

x= [0,4.95,13.1,0,4.95,8.15,0,8.15,13.1]
y = [0.86,0.075,0.11,0.94,0.33,0.69,0.83,0.74,0.38] # 10, not 9, so
#the fit isn't perfect

coef = np.polyfit(x,y,1)
poly1d_fn = np.poly1d(coef)
# poly1d_fn is now a function which takes in x and returns an estimate for y

plt.plot(x,y, 'bo', x, poly1d_fn(x), '--k') #'--k'=black dashed line, 'yo' = yellow circle marker

plt.xlim(0, 15)
plt.ylim(0, 1)
```

```python
p, res, _, _, _ = np.polyfit(x, y, 1, full=True)
```

```python
print(res)
```

```python
print(p)
```
